```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here

fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(10,7), sharex=True, sharey=True)
ax[0].plot(x,tan, label= 'tanX')
ax[1].plot(x,cos, color='red', label='sinX')
ax[2].plot(x,sin,color='green',label='cosX')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here

fig,ax = plt.subplots(figsize=(10,10), sharex = True, sharey=True)
c = ax.plot(x,y,color='b')
ax.set_title('grafik cosX')
akatas=ax.inset_axes([0.8,0.9,0.3,0.3])
akaba=ax.inset_axes([0.8,-0.10,0.3,0.3])
akiba = ax.inset_axes([-0.20,-0.10,0.3,0.3])
akitas = ax.inset_axes([-0.20,0.9,0.3,0.3])

akitas.plot(x,y5,color='y')
akitas.plot(x,y,color='b')

akatas.plot(x,y4,color='black')
akatas.plot(x,y,color='b')

akiba.plot(x,y2,color='red')
akiba.plot(x,y,color='b')

akaba.plot(x,y3,color='green')
akaba.plot(x,y,color='b')
plt.show()
```
