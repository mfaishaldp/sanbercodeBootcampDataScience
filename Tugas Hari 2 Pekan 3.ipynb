Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,10))
ax.plot(x,cos, color='red',label='Cosx')
ax.plot(x,cos3, linestyle='--',color='green',label='3Cosx')
ax.plot(x,sin,linestyle='-.',color='blue',label='Sin(x-90)')
ax.plot(x,sin2,color='yellow',label='2Sinx',marker='x',linestyle='')
ax.legend(loc='center')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
plt.show()
```
