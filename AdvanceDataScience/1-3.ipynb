```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

alamat = "https://en.wikipedia.org/wiki/List_of_brightest_stars"
html = urlopen(alamat)
data = BeautifulSoup(html,"html.parser")

# print(data)

table = data.find("table",{"class" : "wikitable"})
rows = table.findAll("tr")

hasil = []
for row in rows:
    info = []
    for cell in row.findAll(["td","th"]):
        info.append(cell.get_text().strip())
    hasil.append(info)
    
# print(hasil)

df = pd.DataFrame(hasil[1:],columns = hasil[0])
df.set_index('Rank', inplace=True)
df['part 2'][0] = None
df['Distance (ly)'][0] = '0.000015813'
df['Spectral class'][0] = 'G2 V'
df
```
